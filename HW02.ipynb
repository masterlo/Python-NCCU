## 圓: 中心為(3, 5), 半徑為2         

```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
```

```python
θ = np.linspace(-np.pi ,np.pi, 500)
z = 2*np.cos(θ)+2*np.sin(θ)*1j+3+5*1j

ax = plt.gca()
ax.set_aspect('equal')
plt.plot(np.real(z), np.imag(z), '#ff0000')
```

## 橢圓: 中心為(2, 10), 長軸為4.5, 短軸為3

```python
β = np.linspace(-np.pi ,np.pi, 500)
y = 3*np.cos(β)+4.5*np.sin(β)*1j+2+10*1j

ax = plt.gca()
ax.set_aspect('equal')
plt.plot(np.real(y), np.imag(y), '#3366ff')
```
